# 4. Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt
import pandas as pd


plt.rcParams['figure.figsize'] = (8,5)
```

## 1: What percentage of customers have churned?

```python
churn_counts = df['Churn'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    churn_counts,
    labels=churn_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Customer Churn Distribution")
plt.show()
```

### Business Insight

Approximately **25.9%** of customers have churned, while **74.1%** remain with the company.

Although the majority of customers continue using the company's services, losing nearly one out of every four customers is a significant concern. This level of churn can reduce revenue and increase the cost of acquiring new customers.

## 2: Which Contract Type Has the Highest Churn?

```python
contract_churn = pd.crosstab(df['Contract'], df['Churn'])

contract_churn.plot(kind='bar')

plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)
plt.show()
```

### Insight

Customers with **Month-to-Month** contracts show the highest churn.

Long-term contracts have significantly better customer retention, suggesting contract duration plays a major role in reducing churn.

## 3: Does Internet Service Affect Churn?

```python
internet_churn = pd.crosstab(df['InternetService'], df['Churn'])

internet_churn.plot(kind='bar')

plt.title("Churn by Internet Service")
plt.xlabel("Internet Service")
plt.ylabel("Customers")
plt.xticks(rotation=0)
plt.show()
```

Fiber optic customers have the highest churn compared to DSL and customers without internet service.

This may indicate pricing issues, service quality concerns, or unmet customer expectations.

## 4:Does Payment Method Affect Churn?

```python
payment_churn = pd.crosstab(df['PaymentMethod'], df['Churn'])

payment_churn.plot(kind='bar')

plt.title("Churn by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Customers")
plt.xticks(rotation=45)
plt.show()
```

### Business Insight

Customers using **Electronic Check** have the highest churn, while customers using automatic payment methods have the lowest churn. This suggests that payment method is associated with customer retention.

## 5: Electronic check users have noticeably higher churn than customers using automatic payment methods.

## 6: Does Gender Affect Churn?

## Are male and female customers equally likely to churn?

```python
gender_churn = pd.crosstab(df['gender'], df['Churn'])

gender_churn.plot(kind='bar')

plt.title("Gender vs Churn")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)
plt.show()
```

## 7: Do Senior Citizens Churn More?

```python
senior_churn = pd.crosstab(df['SeniorCitizen'], df['Churn'])

senior_churn.plot(kind='bar')

plt.title("Customer Churn by Senior Citizen Status")
plt.xlabel("Senior Citizen")
plt.ylabel("Customers")
plt.xticks([0,1],["No","Yes"], rotation=0)
plt.show()
```

```python
# Churn percentage by Senior Citizen
senior_churn_rate = pd.crosstab(
    df['SeniorCitizen'],
    df['Churn'],
    normalize='index'
) * 100

print(senior_churn_rate)
```

### Business Insight

Senior citizens have a higher churn rate than non-senior customers. This indicates that older customers are more likely to discontinue the company's services.

The company should investigate the reasons behind this trend and consider targeted retention strategies, such as dedicated customer support, loyalty benefits, or customized service plans for senior customers.

## 13: Are High-Paying Customers More Likely to Churn?

```python
plt.figure(figsize=(8,5))

plt.boxplot(
    [df[df['Churn']=="No"]['MonthlyCharges'],
     df[df['Churn']=="Yes"]['MonthlyCharges']],
    labels=["Stayed","Churned"]
)

plt.title("Monthly Charges vs Churn")
plt.ylabel("Monthly Charges")
plt.show()
```

### Business Insight

Customers who churn have a higher average monthly charge than customers who stay with the company. This suggests that higher monthly charges may increase the likelihood of customer churn.

Price-sensitive customers may choose to switch to competitors offering lower prices.

### Business Recommendation

The company should consider offering discounts, loyalty rewards, or better value-added services to customers with high monthly charges to reduce churn.


## 14: Do Long-Term Customers Stay?

```python
plt.figure(figsize=(8,5))

plt.boxplot(
    [df[df['Churn']=="No"]['tenure'],
     df[df['Churn']=="Yes"]['tenure']],
    labels=["Stayed","Churned"]
)

plt.title("Tenure vs Churn")
plt.ylabel("Months")
plt.show()
```

### Business Insight

Customers who stay with the company generally have a much longer tenure than customers who churn. Most customers who leave the company have been with the company for only a short period.

This indicates that new customers are at a higher risk of churning.

### Business Recommendation

The company should focus on improving the experience of new customers through onboarding programs, welcome offers, and regular follow-ups during the first few months to increase customer retention.

2. Contract Type vs Churn
Question

Does the type of contract affect customer churn?

What I found

Customers with Month-to-Month contracts leave much more often than customers with 1-year or 2-year contracts.

Why?

A month-to-month customer has no long-term commitment. If they receive a better offer from another company, they can leave easily.

A customer with a yearly contract is more committed to staying.

Offer discounts or rewards to encourage customers to switch from month-to-month plans to yearly contracts.
